```python
import re
import time
from bs4 import BeautifulSoup
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime

# =========================================================================
# UTILITY HELPER FUNCTION (METADATA EXTRACTOR)
# =========================================================================
def extract_game_metadata(html_content: str) -> dict:
    """Scans the HTML payload to extract tracking UUID, Teams, and Season Year/DateTime."""
    soup = BeautifulSoup(html_content, 'lxml')

    # 1. Pull Tracking UUID from the schedule link
    schedule_link = soup.find('a', attrs={"data-testid": "event-nav-bar-recap"}, href=re.compile(r'/schedule/[a-f0-9\-]{36}'))
    if not schedule_link:
        raise ValueError("Invalid HTML structure. Could not locate GameChanger tracking UUID link with data-testid='event-nav-bar-recap'.")
    game_uuid = re.search(r'/schedule/([a-f0-9\-]{36})', schedule_link['href']).group(1)

    # Extract year from the schedule link's href - this is usually more reliable for the year
    date_match_year = re.search(r'/(\d{4})-(summer|spring|fall|winter)', schedule_link['href'])
    game_year = date_match_year.group(1) if date_match_year else str(datetime.now().year) # Fallback to current year

    # 2. Pull Matchup Slug using 'away-team-name' and 'home-team-name'
    away_team_element = soup.find(attrs={"data-testid": "away-team-name"})
    home_team_element = soup.find(attrs={"data-testid": "home-team-name"})

    team_slug = "Unknown-Matchup"
    if away_team_element and home_team_element:
        away_team_name = away_team_element.get_text(strip=True)
        home_team_name = home_team_element.get_text(strip=True)
        # Clean up names for slug: remove non-alphanumeric, replace spaces with hyphens
        clean_away = re.sub(r'[^a-zA-Z0-9]', '', away_team_name).replace(' ', '-')
        clean_home = re.sub(r'[^a-zA-Z0-9]', '', home_team_name).replace(' ', '-')
        team_slug = f"{clean_away}-vs-{clean_home}"
        # Further clean by replacing multiple hyphens and stripping at ends
        team_slug = re.sub(r'-+', '-', team_slug).strip('-')

    # 3. Pull DateTime Slug using 'event-time' and the extracted game_year
    game_datetime_slug = f"{game_year}_UnknownTime" # Default fallback for full slug
    event_time_element = soup.find(attrs={"data-testid": "event-time"})
    if event_time_element:
        event_time_str_raw = event_time_element.get_text(strip=True)

        # Remove any time range and timezone information for parsing the start time
        # Example: "Sat Jun 13, 9:00 AM - 11:00 AM ET" -> "Sat Jun 13, 9:00 AM"
        # Example: "Fri May 10, 6:00 PM ET" -> "Fri May 10, 6:00 PM"
        match_start_time = re.match(r'^(.*? (?:AM|PM))', event_time_str_raw)
        event_time_clean_for_parsing = match_start_time.group(1).strip() if match_start_time else event_time_str_raw.strip()

        parsed_date = None
        try:
            # Construct a full datetime string with the extracted game_year
            full_date_time_str = f"{event_time_clean_for_parsing} {game_year}"
            # Attempt to parse common date format 'Mon, Mar 13, 9:00 AM 2023' or 'Sat Jun 13, 9:00 AM 2023'
            parsed_date = datetime.strptime(full_date_time_str, '%a %b %d, %I:%M %p %Y')
        except ValueError:
            # Fallback to previously successful formats if this specific one fails
            try:
                # Attempt to parse common date format 'Mon, Mar 1, 2023 10:00 AM' (original with year, might not have 'ET')
                parsed_date = datetime.strptime(event_time_str_raw, '%a, %b %d, %Y %I:%M %p')
            except ValueError:
                try:
                    # Attempt to parse common date format 'YYYY-MM-DD HH:MM:SS'
                    parsed_date = datetime.strptime(event_time_str_raw, '%Y-%m-%d %H:%M:%S')
                except ValueError:
                    print(f"Warning: Could not parse event time '{event_time_str_raw}'. Using default year for datetime slug.")

        if parsed_date:
            game_datetime_slug = parsed_date.strftime('%Y%m%d_%H%M')
        else:
            game_datetime_slug = f"{game_year}_UnknownTime" # If all parsing fails, use extracted year with UnknownTime

    return {"uuid": game_uuid, "teams": team_slug, "year": game_year, "datetime_slug": game_datetime_slug}

# =========================================================================
# UI ELEMENT INITIALIZATIONS
# =========================================================================
box_input = widgets.Textarea(
    value='',
    placeholder='Paste entire Box Score rendered HTML source code here...',
    description='Box HTML:',
    layout=widgets.Layout(width='98%', height='200px')
)

plays_input = widgets.Textarea(
    value='',
    placeholder='Paste entire Play-by-Play rendered HTML source code here...',
    description='Plays HTML:',
    layout=widgets.Layout(width='98%', height='200px')
)

run_button = widgets.Button(
    description='Run Ingestion Pipeline',
    button_style='success', # Creates a clean green execution button
    tooltip='Click to process payloads and output unique staging files',
    icon='check-circle',
    layout=widgets.Layout(width='300px', height='45px', margin='15px 0px 0px 0px')
)

output_console = widgets.Output()

# =========================================================================
# PIPELINE BUTTON CLICK TRIGGER TRIGGER
# =========================================================================
def on_button_clicked(b):
    with output_console:
        clear_output() # Clears logs from previous click runs

        box_text = box_input.value.strip()
        plays_text = plays_input.value.strip()

        # Validation checks
        if not box_text or not plays_text:
            print("🛑 PIPELINE FAILURE: Both text fields must be populated before processing.")
            return

        print("⚡ Ingestion Engine Started...")
        timestamp = int(time.time())

        # 1. Process and save the Box Score file
        try:
            box_meta = extract_game_metadata(box_text)
            box_filename = f"{box_meta['datetime_slug']}_{box_meta['teams']}_{box_meta['uuid']}_box.html"
            with open(box_filename, "w", encoding="utf-8") as f:
                f.write(box_text)
            print(f"✅ Box Score Staged: {box_filename}")
        except Exception as e:
            box_filename = f"fallback_game_{timestamp}_box.html"
            with open(box_filename, "w", encoding="utf-8") as f:
                f.write(box_text)
            print(f"⚠️ Box Metadata failed. Saved to emergency fallback: {box_filename}. Error: {e}")

        # 2. Process and save the Play-by-Play file
        try:
            plays_meta = extract_game_metadata(plays_text)
            plays_filename = f"{plays_meta['datetime_slug']}_{plays_meta['teams']}_{plays_meta['uuid']}_plays.html"
            with open(plays_filename, "w", encoding="utf-8") as f:
                f.write(plays_text)
            print(f"✅ Play-by-Play Staged: {plays_filename}")
        except Exception as e:
            plays_filename = f"fallback_game_{timestamp}_plays.html"
            with open(plays_filename, "w", encoding="utf-8") as f:
                f.write(box_text)
            print(f"⚠️ Plays Metadata failed. Saved to emergency fallback: {plays_filename}. Error: {e}")

        print("\n🚀 Pipeline finished successfully! Files are now ready in your workspace storage sidebar.")

# Bind the function to the UI button element
run_button.on_click(on_button_clicked)

# =========================================================================
# RENDER VISUAL LAYOUT CONTAINER
# =========================================================================
panel_ui = widgets.VBox([
    widgets.HTML(value="<h2>GameChanger Scraping & Ingestion Panel</h2>"),
    widgets.HTML(value="<p>Extract HTML via your browser DevTools Console using `copy(document.documentElement.outerHTML);` and paste below:</p>"),
    box_input,
    widgets.HTML(value="<div style='margin-top:10px;'></div>"),
    plays_input,
    run_button,
    widgets.HTML(value="<hr style='border-top: 1px solid #ccc; margin-top:15px;'/>"),
    widgets.HTML(value="<h4>Pipeline Execution Logs:</h4>"),
    output_console
])

display(panel_ui)
```
